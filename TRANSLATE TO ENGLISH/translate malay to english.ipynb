```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#search term

query = "('Celcom murah OR Celcom laju OR Celcom slow OR Celcom bagus OR Celcom mahal OR Celcom  tiada line' )  until:2023-03-31 since:2019-03-31"

tweets = []
limit = 150


for tweet in sntwitter.TwitterHashtagScraper(query).get_items():
    
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.url, tweet.user.username, tweet.sourceLabel, tweet.user.location, tweet.content])
        
df = pd.DataFrame(tweets, columns=['Date', 'TweetURL','User', 'Source', 'Location', 'Text'])
```
